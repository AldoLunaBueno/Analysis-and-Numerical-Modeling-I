```python
from cmath import pi
from numpy import sin,cos,tan

def sex_rad(angle):
    return (angle/180)*pi
def rad_sex(angle):
    return angle/pi*180
eps = 1E-5
maxIter = 100
l = 2.2
h = 1.2
D = 0.75
beta1= sex_rad(11.5)

A = l*sin(beta1)
B = l*cos(beta1)
C = (h+0.5*D)*sin(beta1) - 0.5*D*tan(beta1)
E = (h+0.5*D)*cos(beta1)-0.5*D

f = lambda x: A*sin(x)*cos(x)+B*sin(x)**2-C*cos(x)-E*sin(x)



x0 = 0
x1 = pi
f0 = f(x0)
f1 = f(x1)
i=0
while i<maxIter and abs(f1)>=eps:
    h = -f1*(x1-x0)/(f1-f0)
    x0 = x1
    x1 = x1 +h
    f0 = f1
    f1 = f(x1)
    i+=1
    print(f'x1={x1:5.3f} f(x1)={f1:5.3f}')
if abs(f1) >= eps:
    print(f'Metodo no converge')
else:
    print(f'Solucion c={rad_sex(x1)}')
    print(f'Numero de iteraciones={i}')
    
    
```
